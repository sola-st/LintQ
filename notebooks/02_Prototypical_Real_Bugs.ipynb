## Rule 1: ClassicalRegister + measure_all
*(Zhao et al. Q-SE Workshop 2022)*

LINT: measure_all not using your predefined classical register.
measure_all risks of creating another classical register, beside 
the one you initialized.

```python
test_positive_code = """from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer

# Create a Quantum Register with 2 qubits.
q = QuantumRegister(2)
# Create a Classical Register with 2 bits.
c = ClassicalRegister(2)
# Create a Quantum Circuit
qc = QuantumCircuit(q, c)
qc.h(q[0])
qc.cx(q[0], q[1])
qc.measure_all()
# Execute the circuit on qasm simulator
job = execute(qc, Aer.get_backend('qasm_simulator'), shots=1000)
# Grab results from the job
result = job.result()
# Returns counts
counts = result.get_counts(qc)

print(counts)"""
```

```python
def prettify(ast_tree_str, indent=4):
    ret = []
    stack = []
    in_string = False
    curr_indent = 0

    for i in range(len(ast_tree_str)):
        char = ast_tree_str[i]
        if in_string and char != '\'' and char != '"':
            ret.append(char)
        elif char == '(' or char == '[':
            ret.append(char)

            if i < len(ast_tree_str) - 1:
                next_char = ast_tree_str[i+1]
                if next_char == ')' or next_char == ']':
                    curr_indent += indent
                    stack.append(char)
                    continue

            print(''.join(ret))
            ret.clear()

            curr_indent += indent
            ret.append(' ' * curr_indent)
            stack.append(char)
        elif char == ',':
            ret.append(char)

            print(''.join(ret))
            ret.clear()
            ret.append(' ' * curr_indent)
        elif char == ')' or char == ']':
            ret.append(char)
            curr_indent -= indent
            stack.pop()
        elif char == '\'' or char == '"':

            if (len(ret) > 0 and ret[-1] == '\\') or (in_string and stack[-1] != char):
                ret.append(char)
                continue

            if len(stack) > 0 and stack[-1] == char:
                ret.append(char)
                in_string = False
                stack.pop()
                continue

            in_string = True
            ret.append(char)
            stack.append(char)
        elif char == ' ':
            pass
        else:
            ret.append(char)

    print(''.join(ret))
```

```python
tree = ast.parse(test_positive_code)
ast_tree_str = ast.dump(tree)
prettify(ast_tree_str)
```

```python
from pandarallel import pandarallel
import ast
import astor
pandarallel.initialize(progress_bar=True)

def check_bug(code:str):
    """Isolate al the code using the `measure_all` function call."""
    tree = ast.parse(code)
    for node in ast.walk(tree):
        if isinstance(node, ast.Call):
            if isinstance(node.func, ast.Attribute):
                if node.func.attr == 'measure_all':
                    return True
    return False

check_bug(test_positive_code)
```

```python
def safe_check_bug(code:str):
    try:
        return check_bug(code)
    except:
        return False

df_parsable_programs['bug_01'] = \
    df_parsable_programs['file_content_code_only'].apply(
        lambda c: safe_check_bug(c)
)
df_parsable_programs.head()
```

## Rule 4: Repeated Measurement 
*(Zhao et al. Q-SE Workshop 2022)*

LINT: redundant measurement. When measuring the same qubit
over and over in tha same loop without any operation in between.

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import execute, Aer

def get_circuit(n):
    qreg = QuantumRegister(1)
    creg = ClassicalRegister(n)
    circ = QuantumCircuit(qreg, creg)
    
    for i in range(n):
        circ.measure(qreg[0], creg[i])
    return circ

aer = Aer.get_backend('qasm_simulator')
circ65 = get_circuit(4)
counts = execute(circ65, aer).result().get_counts()
print(counts)
# {'0000': 1024}
```

## Rule 5: Unused Qubit 
*(new idea)*

LINT: Unused bit, You are not using the full register, 
you can downsize your algorithm

```python
tq = QuantumRegister(3)
tc0 = ClassicalRegister(1)
tc1 = ClassicalRegister(1)
tc2 = ClassicalRegister(1)
teleport = QuantumCircuit(tq, tc0,tc1,tc2)
teleport.h(tq[1])
teleport.cx(tq[1], tq[2])
teleport.ry(np.pi/4,tq[0])
teleport.cx(tq[0], tq[1])
teleport.h(tq[0])
teleport.barrier()
teleport.measure(tq[0], tc0[0])
teleport.measure(tq[1], tc1[0])
teleport.cx(tq[1], tq[2])
teleport.cz(tq[0], tq[2])
teleport.measure(tq[2], tc2[0])

teleport.draw()
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import execute, Aer
tq = QuantumRegister(3)
tc0 = ClassicalRegister(1)
tc1 = ClassicalRegister(1)
tc2 = ClassicalRegister(1)
teleport = QuantumCircuit(tq, tc0,tc1,tc2)
teleport.h(tq[1])
teleport.cx(tq[1], tq[2])
teleport.ry(np.pi/4,tq[0])
teleport.cx(tq[0], tq[1])
teleport.h(tq[0])
teleport.barrier()
teleport.measure(tq[0], tc0[0])
teleport.measure(tq[1], tc1[0])
teleport.cx(tq[1], tq[2])  # BUG
teleport.cz(tq[0], tq[2])  # BUG
teleport.measure(tq[2], tc2[0]) 

```
